```python
df2=df1[df1.TotalCharges!=' ']
```

```python
df2.TotalCharges=pd.to_numeric(df2.TotalCharges)
```

```python
df2.shape
```

```python
df2.dtypes
```

```python
for col in df2:
    print(f'{col} : {df2[col].unique()}')
```

```python
df2.TotalCharges.values
```

```python
df3=pd.get_dummies(data = df2,columns = ['Contract','PaymentMethod','InternetService'])
df3.dtypes
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df3[cols_to_scale] = scaler.fit_transform(df3[cols_to_scale])
```

```python
for col in df3:
    print(f'{col} : {df3[col].unique()}')
```

```python
X = df3.drop('Churn',axis='columns')
y = df3['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```
